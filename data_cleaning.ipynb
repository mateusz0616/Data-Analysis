```python
df['author']=df['author'].str.replace('Writtenby:','').str.strip()
```

```python
full_name = pd.Series(name="full_name")

def clear_name():
    
    for i in df['author']:
        formatted_name = re.sub(r"([A-Z])", r" \1", i).strip()
        full_name.loc[len(full_name)] = formatted_name
        
        
    
clear_name() 
print(full_name)     
```

```python
df['author']=full_name
```

```python
df['author']=df['author'].str.replace('. ','.')
```

```python
df['narrator']=df['narrator'].str.replace('Narratedby:','')
```

```python
df['narrator']
```

```python
narraotr_name = pd.Series(name="name")

def clear_narator():
    
    for i in df['narrator']:
        changed = re.sub(r"([A-Z])", r" \1", i).strip()
        narraotr_name.loc[len(narraotr_name)] = changed
        
        
    
clear_narator() 
df['narrator']=narraotr_name
```

```python
df
```

```python
df['releasedate']=pd.to_datetime(df['releasedate'])
```

```python
df['language']=df['language'].str.capitalize()
```

```python
df['price']=df['price'].str.replace('Free','0')
df['price']=df['price'].apply(lambda x: float(x.replace(',', '')))
```

```python
df['Stars']=df['stars'].str.split().str[0]
df['Stars']=df['Stars'].str.replace('Not','Not rated yet')
```

```python
df['Reviews']=df['stars'].str.split().str[4].str.strip('stars')
```

```python
df=df.drop('stars',axis=1)
```

```python
df['Reviews'].fillna(0,inplace=True)
```

```python

df['time']=df['time'].str.replace(' hrs and ',":").str.strip('mins min hrs').str.replace(' hr and ',':')
```

```python
df=df.loc[~df['time'].str.contains('Less')]
```

```python
df
```

```python
hour_to_min=pd.Series(name='time')
hour_to_min=df['time'].str.split(':').str[0].astype(int)
hour_to_min=hour_to_min*60
```

```python
add_mins=pd.Series(name='mins')
add_mins=df['time'].str.split(':').str[-1].astype(int)
```

```python
df['time']=hour_to_min+add_mins
df['time']=pd.to_timedelta(df['time'], unit='m')
```
